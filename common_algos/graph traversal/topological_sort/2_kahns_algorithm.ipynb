## Kahn's Algorithm for Topological Sort

- Kahn's algorithm is essentially breadth first search on a graph, with the additional overhead of tracking the in-degrees of every node in the graph

### How does Kahn's Algorithm work?

- Suppose we have a list of packages, where every package either depends on some other package, or has no dependencies

- Let's define the number of dependencies a package has as its "In-Degree"

- Kahn's algorithm computes the in-degree of every package. 
    - For packages with no dependencies (in-degree = 0), add them to a queue.
    - Pop from the queue to install packages. For every package that depends on the installed package, decrement their in-degree by 1
    - When a package hits in-degree of 0, add it to the queue

- Note that this is by construction a **directed** graph. Kahn's algorithm will not work on undirected graphs, because no node in an undirected graph can have an in-degree of 0 as long as there is an edge

- Note that this approach only works under the assumption that the underlying dependencies do not form a cycle. Why? 
    - Imagine a simple A --> B --> C --> A cycle
    - Computing in degrees, we get {A: 1, B: 1, C: 1}. We are not able to add anything to a queue!
    - All cycles will have this problem; there will come a point where we cannot add anything to the queue, because nothing has an in-degree of 0s

### Implementation

```python
from graphviz import Digraph

dag_graph = {
    0: [],
    1: [],
    2: [3],
    3: [1],
    4: [0, 1],
    5: [0, 2]
}

def visualize_kahns_graph(graph):
    dot = Digraph(comment='DAG for Kahn\'s Algorithm')
    dot.attr(rankdir='LR')  # Left to Right flow
    
    # 2. Add nodes and edges
    for node, neighbors in graph.items():
        # Label nodes with their ID and their current In-Degree
        label = f"Node: {node}"
        
        # Highlight source nodes (In-Degree 0) in gold
        color = 'lightblue'
        
        dot.node(str(node), label=label, style='filled', fillcolor=color)
        
        for neighbor in neighbors:
            dot.edge(str(node), str(neighbor))

    return dot

visualize_kahns_graph(dag_graph)
```

- The graph above simply means, for example, that node 0 depends on nodes 4 and 5. So node 0 has an in-degree of 2

```python
from collections import deque

def compute_in_degree(graph):
    in_degrees = {k: 0 for k in graph.keys()}
    for k, v in graph.items():
        for elem in v:
            in_degrees[elem] += 1
    return in_degrees


def kahns_algorithm(graph):
    in_degree = compute_in_degree(graph)
    queue = deque()
    visited = set()
    for k, v in in_degree.items():
        if v == 0:
            queue.append(k)
            visited.add(k)
    
    while queue:
        curr_node = queue.popleft()
        print(f"Visiting {curr_node=}")
        for adj_node in graph[curr_node]:
            if adj_node in visited:
                raise RuntimeError("NODE ALREADY VISITED, CYCLE DETECTED")
            in_degree[adj_node] -= 1
            if in_degree[adj_node] == 0:
                queue.append(adj_node)
                visited.add(adj_node)

    if len(visited) != len(in_degree):
        raise RuntimeError("LEFTOVER UNVISITED NODES, CYCLE DETECTED")

    return 

compute_in_degree(dag_graph)
kahns_algorithm(dag_graph)
```
