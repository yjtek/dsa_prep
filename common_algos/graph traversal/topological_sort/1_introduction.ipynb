## Topological Sort: Motivation

- In typical sorting, you want to resolve the order of some set of objects that relate to each other in some predefined manner. 
    - For example, in numeric values, 1 < 2 < 3.

- For topological sorting, we are not interested in the exact values of our objects, but in their relationship with each other.
    - That is, we are interested in the network/graph structure of the objects, not their exact values. 
    - For example, if I have 10 different scheduled jobs that are all mutually dependent, how should I schedule them such that I run all leaf dependencies first?
    - Or in jargon, how do I create a **dynamic acyclic graph** (DAG)

- So in topological sort, though it is a sort of sorting, is a little more involved.  
    - Because in typical sorting, we expect relationships to be transitive. That is, 1 < 2 must mean that 1 < 3. 
    - But DAGs do not obey transitivity. If A --> B --> D, and C --> D, that does not mean that A and B have any relation to C 
    - In fact, so long as there are no cycles, any structure is possible.

- We'll cover 2 common algorithms to perform this topological sort
    - Kahn's Algorithm
    - Depth First Search (DFS)
