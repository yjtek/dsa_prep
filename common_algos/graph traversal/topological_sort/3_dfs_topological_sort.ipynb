## DFS Topological Sort

- Kahn's algorithm is the canonical approach to solving the topological sort problem

- DFS is technically possible, with some major caveats
    - You need to maintain 3 states for every node; is it unvisited, visited, or VISITING
    - This is needed because you need to know if, upon seeing a node, you have encountered this on the current recursion path, or on a different path.
    - For example, 0 --> 1 --> 2 --> 0 
    - When you see the second 0, you need a way to tell if it is on the same path (and thus cycle) or a different path due to DFS backtracking

- Kahn's algorithm is far clearer and far more canonical

- We will explore implementation with the graph below

```python
from graphviz import Digraph

dag_graph = {
    0: [],
    1: [],
    2: [3],
    3: [1],
    4: [0, 1],
    5: [0, 2]
}

def visualize_kahns_graph(graph):
    dot = Digraph(comment='DAG for Kahn\'s Algorithm')
    dot.attr(rankdir='LR')  # Left to Right flow
    
    # 2. Add nodes and edges
    for node, neighbors in graph.items():
        # Label nodes with their ID and their current In-Degree
        label = f"Node: {node}"
        
        # Highlight source nodes (In-Degree 0) in gold
        color = 'lightblue'
        
        dot.node(str(node), label=label, style='filled', fillcolor=color)
        
        for neighbor in neighbors:
            dot.edge(str(node), str(neighbor))

    return dot

visualize_kahns_graph(dag_graph)
```

### Implementation

```python
def dfs_topo_sort_with_cycle_detection(graph):
    
    ## Track the nodes that we already visited 
    visited = set()

    ## Track the current path we are following
    rec_stack = set()
    
    ## Once our traversal reaches leaf nodes, append to this. 
    order = []

    def check_visit_success(node):
        ## Check if node is already in rec stack, if yes, cycle detected, return false
        if node in rec_stack:
            return False

        ## Check if node is already visited
        if node in visited:
            return True

        ## If not in rec stack, AND not visited yet, add to rec stack...
        rec_stack.add(node)

        ## ...and visit neighbours
        for neighbor in graph.get(node, []):
            if not check_visit_success(neighbor):
                return False

        ## Once successfully visited neighbours, remove node from traversal stack
        rec_stack.remove(node)

        ## Indicate that we have successfully visited the node
        visited.add(node)

        ## Add node to the update stack. Note that this node will come AFTER its dependencies
        order.append(node)
        return True

    ## For every node in the graph
    for node in graph:

        ## Visit unvisited nodes
        if node not in visited:
            if not check_visit_success(node):
                return "Cycle Detected! Topological Sort impossible."
    
    ## Since nodes are appended AFTER their dependencies, we essentially have our nodes in reverse 
    ## update order. So return the reversed list
    return order[::-1]

dfs_topo_sort_with_cycle_detection(dag_graph)
```
