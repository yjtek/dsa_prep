## Breadth-First Search

- The idea of BFS is that, in a tree, I search level by level from root down to node

- Let's suppose we have 3 graphs below; an undirected graph, a directed acyclic graph, and a directed cyclic graph

- BFS works for all of them!

```python
undirected_graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 4, 5],
    3: [1, 5],
    4: [1, 2, 5],
    5: [3, 4]
}

directed_acyclic_graph = {
    0: [1, 2],
    1: [3],
    2: [3, 4, 5],
    3: [5],
    4: [5],
    5: []
}

cyclic_graph = {
    0: [1],
    1: [2],
    2: [3, 4],
    3: [1],
    4: [0]
}
```

```python
from graphviz import Digraph, Graph

def visualize_structure(adj_list, title, directed=True):
    # Choose the correct class based on graph type
    dot = Digraph(name=title) if directed else Graph(name=title)
    
    dot.attr(label=f"\n\n{title}", fontsize='20')
    dot.attr(layout='dot') # 'dot' is great for hierarchical structures like DAGs
    dot.node_attr.update(shape='circle', style='filled', fillcolor='lightyellow')

    edges_seen = set()
    for node, neighbors in adj_list.items():
        dot.node(str(node))
        for neighbor in neighbors:
            # For undirected graphs, we avoid drawing the same edge twice
            edge = tuple(sorted((str(node), str(neighbor))))
            if not directed and edge in edges_seen:
                continue
            
            dot.edge(str(node), str(neighbor))
            edges_seen.add(edge)
    return dot

display(visualize_structure(undirected_graph, 'undirected_graph', directed=False))
display(visualize_structure(directed_acyclic_graph, 'dag', directed=True))
display(visualize_structure(cyclic_graph, 'cyclic', directed=True))
```

- BFS ensures that traversal happens from root downwards

### BFS Traversal

- In BFS, we use the adjacency map of the graph $A$, a queue $q$, and an array $V$ to record when nodes have been visited
    - Init `V = set([0])`, indicating that we visit the first node
    - Init `q = [0]`, indicating starting node.
    - popleft from $q$ to get starting node 0, and query $A$ for adjacencies
    - For each adjacency, append to queue and add to visited
    - Loop

- In the end, this ensures that we will visit every node in the graph according to their distance from the first node

```python
from collections import deque

def bfs(graph, start_node=0):
    visited = set([start_node])
    queue = deque([start_node])
    while queue:
        curr_node = queue.popleft()
        print(f'visiting {curr_node=}')
        for adj_node in graph[curr_node]:
            if adj_node not in visited:
                queue.append(adj_node)
                visited.add(adj_node)

# bfs(undirected_graph)
# bfs(directed_acyclic_graph)
# bfs(cyclic_graph)
```

- We can see most clearly how the traversal works in undirected graphs by simply starting from a different node. You can see quite clearly that nodes directly connected to the starting node are visited first, followed by those further away

```python
bfs(undirected_graph, 4)
```
