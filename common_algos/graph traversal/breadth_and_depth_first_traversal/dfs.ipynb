## Depth-First Search

- The idea of DFS is that, for any graph, I follow each path to the end before going back up to visit previous "forks" in the path

- Let's suppose we have 3 graphs below; an undirected graph, a directed acyclic graph, and a directed cyclic graph

- DFS works for all of them!

```python
undirected_graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 4, 5],
    3: [1, 5],
    4: [1, 2, 5],
    5: [3, 4]
}

directed_acyclic_graph = {
    0: [1, 2],
    1: [3],
    2: [3, 4, 5],
    3: [5],
    4: [5],
    5: []
}

cyclic_graph = {
    0: [1],
    1: [2],
    2: [3, 4],
    3: [1],
    4: [0]
}
```

```python
from graphviz import Digraph, Graph

def visualize_structure(adj_list, title, directed=True):
    # Choose the correct class based on graph type
    dot = Digraph(name=title) if directed else Graph(name=title)
    
    dot.attr(label=f"\n\n{title}", fontsize='20')
    dot.attr(layout='dot') # 'dot' is great for hierarchical structures like DAGs
    dot.node_attr.update(shape='circle', style='filled', fillcolor='lightyellow')

    edges_seen = set()
    for node, neighbors in adj_list.items():
        dot.node(str(node))
        for neighbor in neighbors:
            # For undirected graphs, we avoid drawing the same edge twice
            edge = tuple(sorted((str(node), str(neighbor))))
            if not directed and edge in edges_seen:
                continue
            
            dot.edge(str(node), str(neighbor))
            edges_seen.add(edge)
    return dot

display(visualize_structure(undirected_graph, 'undirected_graph', directed=False))
display(visualize_structure(directed_acyclic_graph, 'dag', directed=True))
display(visualize_structure(cyclic_graph, 'cyclic', directed=True))
```

### DFS Traversal

- In DFS, we use the adjacency map of the graph $A$, a stack $s$, and an array $V$ to record when nodes have been visited
    - Init `V = set([0])`, indicating that we visit the first node
    - Init `q = [0]`, indicating starting node.
    - pop from $s$ to get starting node 0, and query $A$ for adjacencies 
    - For each adjacency, add to stack and add to visited
    - Loop

- The stack structure ensures that we keep visiting deeper nodes first, until we get to the root

```python
def dfs(graph, start_node=0):
    visited = set([start_node])
    stack = [start_node]
    while stack:
        curr_node = stack.pop()
        print(f'visiting {curr_node=}')
        for adj_node in graph[curr_node]:
            if adj_node not in visited:
                stack.append(adj_node)
                visited.add(adj_node)

dfs(undirected_graph)
# dfs(directed_acyclic_graph)
# dfs(cyclic_graph)
```

```python
dfs(undirected_graph, 4)
```

- It is also worth noting that DFS does not always require an explicit stack. A cheap way is to use the call stack via recursion instead. But there is no benefit to doing this besides looking smart; explicit stack is much clearer in intent IMO

```python
def dfs_recursive(graph, curr_node=0, visited=None):
    if not visited:
        visited = set()

    visited.add(curr_node)
    print(f'visiting {curr_node=}')
    
    for adj_node in graph[curr_node]:
        if adj_node not in visited:
            dfs_recursive(graph, adj_node, visited)

dfs_recursive(undirected_graph, curr_node=0)
```
