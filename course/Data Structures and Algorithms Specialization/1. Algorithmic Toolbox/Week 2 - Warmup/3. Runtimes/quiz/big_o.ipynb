```python
import matplotlib.pyplot as plt
import numpy as np
n = 1e10
jumps = n/1000
plt.plot(np.arange(0, n, jumps), 5 ** np.emath.logn(2, np.arange(0, n, jumps)), label='5**log(n)')
plt.plot(np.arange(0, n, jumps), np.arange(0, n, jumps)**5, label='n**5')
plt.legend()
```

```python
n = 1e10
jumps = n/1000
plt.plot(np.arange(0, n, jumps), 2 ** (3 * np.emath.logn(2, np.arange(0, n, jumps))), label='2**log(n)')
plt.plot(np.arange(0, n, jumps), np.arange(0, n, jumps)**5, label='n**5')
plt.legend()
```
