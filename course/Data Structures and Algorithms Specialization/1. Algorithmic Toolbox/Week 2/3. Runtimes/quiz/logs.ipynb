- 1 and 2 are by definition
- 3 is because 
$$\begin{aligned}
    n^{\log_a(b)} &= b^{\log_a(n)} \\
    \log_a(b) \cdot \log_n(n) &= \log_a(n) \cdot \log_n(b) \\
    \log_a(b) &= \log_n(n) / \log_n(a) \cdot \log_n(b) / \log_n(n) \\
    \log_a(b) &= \log_n(b) / \log_n(a) \\
    \log_a(b) &= \log_a(b) 
\end{aligned}$$
- 4 is because 
$$\begin{aligned}
    \log_a(n) \cdot \log_b(a) &= \log(n) / \log(a) \cdot \log(a) / \log(b) \\
    &= \log(n) / \log(b) \\
    &= \log_b(n)
\end{aligned}$$
