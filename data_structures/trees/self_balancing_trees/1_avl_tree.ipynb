## AVL Tree

- AVL (named after Adelson-Velsky and Landis) is the first-ever invented self-balancing binary search tree.
    - That is, this guarantees that the difference in heigh between left and right subtrees is at most 1

- This answers the question: How do we ensure a binary tree is balanced in memory?

### Motivation for Self Balancing Trees

- Let's suppose we have the following values that we try to insert into a binary search tree (BST): [5,4,3,2,1]
    - Insert 5, build 1 node
    - Insert 4, create new node on the left of the root
    - Insert 3, create new node on the left of left of root
    - etc.

- This is a degenerate BST where every node has 1 child
    - In such a tree, we completely lose the best feature of a BST, which is the $O(\log N)$ search, insertion, and deletion
    - Why? Because it basically becomes a linked list; all numbers are just linked to 1 other number
    - So to search/insert/delete for a number, we have to traverse the whole chain, which is $O(N)$ search 

- Thus, the "binariness" of a BST is extremely important to our ability to maintain this $O(\log N)$ performance!    


- How do we quantify a tree's "binariness"?
    - Simple! A tree is binary if the tree is approximately height balanced when branching from any given node
    - This can be quantified by something known as the **Balance Factor (BF)** of the tree
    - In an AVL tree, it must be true that $BF \in \{-1, 0, 1\}$

$$\begin{aligned}
    BF(\text{node}) &= \text{Height}(\text{Left Subtree}) - \text{Height}(\text{Right Subtree})
\end{aligned}$$



### How does a tree "self-balance"?

- We now know what it means for a tree to be balanced $BF \in \{-1, 0, 1\}$. How then should we get the tree to balance itself?

- This is done via what is known as **Rotations**

- There are 4 types of rotations, and these are best demonstrated by example
    - Left-Left (LL) Rotation
    - Right-Right (RR) Rotation
    - Left-Right (LR) Rotation
    - Right-Left (RL) Rotation

    - Looking at 50, we can see that the left subtree has height of 3, and the right has a height of 1, leading to $BF = 3 - 1 = 2$

- In an AVL tree, we always rotate the from the imbalance closest to the leaf nodes
    - Therefore, in this case, we rotate 50 first

    - This is done by rotating the entire chain of 10-20-30-50-60 rightwards, so that 30 is the new root
    - 20 remains the left child of 30, and 50 is the right
    - 40 is now an orphan, and so will be transplanted to the new left child of 50

```python
n1.left = n4 # replace 80's left with 30
n2.left = n4.right # transplant 40 to 50
n4.right = n2 # set 50 to the right child of 30
display_visual(n1)
```

#### Checking for balance

- How do we check if a tree is balanced?

- It may be tempting to say that if the root node of a tree is balanced, the whole tree is balanced. This is **wrong**. 

- Consider this tree; 50 is balanced, but early the left and right subtrees are not

```python
t1=MyNode(50)
t1.add_new_node(MyNode(40))
t1.add_new_node(MyNode(30))
t1.add_new_node(MyNode(20))
t1.add_new_node(MyNode(60))
t1.add_new_node(MyNode(70))
t1.add_new_node(MyNode(80))
display_visual(t1)
```

- Therefore, to check for balance, we need to check more than the root node. 

- Our next instinct may be to say that, since it is possible for subtrees to be imbalanced when the root is balanced, we need to check ALL nodes. This, again, is a mistake

- Let's think about this step by step; 
    - An AVL tree can be assumed to be balanced at the point when we have a new node to add (by definition)
    - Once I add a new node, I may potentially unbalance the tree. 
    - HOWEVER, the nodes whose balance I can affect must be an ancestor of the node I just added
    - For example, it is not possible that I add a new node on my left subtree, but end up unbalancing nodes on my right subtree. That does not make sense, since my node is not on the path of any of the subtrees on the right

- Therefore, to check if the new addition breaks that tree's balance, we simply need to trace through all the ancestors of the node we just added!

- What is the best way to check for the balance of ancestors? 2 ways

- We add a pointer to the parent node of every node, and trace backwards from each new node added. This is NOT preferred, because it complicates any rotation that we do. Now, instead of just dealing with left and right child, you need to deal with the parent pointers also

- Instead of adding a pointer, we use **recursion** to check the balance along the entire chain! How?
    - First, we add the new node to the tree as per normal. This involves a recursive call to `add_new_node`; if our left/right children are already assigned, we recursively go down one level to attempt to assign the new node to the child of current node's children
    - After we assign, we check the balance of the current node we are at. 
        - Note the importance of this step! Because of the previous recursion, we perform this check for every ancestor where the new node is added!
    - Because of this recursion, we automatically perform the necessary rotations on imbalanced subtrees
        - Since rotations are guaranteed to create balanced trees, and we perform rotations from the bottom up, the final result MUST be a balanced tree
        - It is impossible to create an imbalanced subtree as the result of a rotation!

```python
def add_new_node(self, value: int):
    ## 1. Add new nodes
    if value < self.value:
        if self.left:
            self.left = self.left.add_new_node(value)
        else:
            self.left = MyNode(value)
    elif value > self.value:
        if self.right:
            self.right = self.right.add_new_node(value)
        else:
            self.right = MyNode(value)
    else:
        raise ValueError("No identical values in tree by assumption")
    
    ## 2. Check for balance
    balance_factor = self.get_balance_factor()

    ## 3. Perform rotations
    if bf > 1: # Left Heavy
        if value < self.left.value: # LL Case
            return self._rotate_right()
        else: # LR Case
            self.left = self.left._rotate_left()
            return self._rotate_right()
    
    if bf < -1: # Right Heavy
        if value > self.right.value: # RR Case
            return self._rotate_left()
        else: # RL Case
            self.right = self.right._rotate_right()
            return self._rotate_left()

    return self
```
