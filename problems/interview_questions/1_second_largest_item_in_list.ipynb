## Problem: Given a list of arbitrary size N, how would you find the second largest item in the list

### Naive: Sort 

- Sort the list with worst case $O(N \log N)$ time and $O(1)$ space (depending on sorting algorithm)
- Pop second item in $O(1)$ time

### Two Pointers

- Maintain 2 pointers, `A` for the largest value observed, `B` for the second largest value observed
- Do one pass through the list. At each element:
    - if value is larger than both `A, B = value, B`
    - if value is smaller than `A` and larger than `B`, `B = value`
    - if value is smaller than both, continue
    - return B at the end
- **Complexity:**
    - 1 pass, so $O(N)$ time, and $O(1)$ space complexity because no copy

- Addition: How does your solution change if there are 1 billion items in the list? (i.e. what if N is large) 
    - Use a Min-Heap of size $K$

- Addition: How does your solution change we want to find the millionth largest item in the list? (i.e. what if K is large) 
    - Use a Min-Heap of size $K$

- Addition: How does your solution handle duplicates?
    - If we consider duplicates as "second largest" (i.e. second largest of [10,10,9] is 10), we proceed as per normal
    - If we want "unique second largest", then we need to add a comparison to see if each new value is equal to our pointers. If it is, ignore it

### Max-Heap

- Using the list provided, create a max heap.
- Then pop the largest value, and heapify. 
- The second pop gives us the desired value

- **Complexity:**
    - Creating the heap takes $O(N)$ time
    - The first pop, and heapify again takes another $O(\log N)$ time
        - This assumes we heapify downwards. Heapifying upwards takes $O(N \log N)$, because we heapify all the root nodes in the heap
    - The second pop takes another $O(\log N)$ assuming we want to retain the heap structure.
    - So overall this takes $O(N + 2 \log N) \approx O(N)$ time
    - BUT because we need to maintain a heap structure, this requires another $O(N)$ space
- Computational overhead: 
    - Heap has larger overhead
    - In the 2 pointer solution, there is at most $2 * N$ comparisons. In the best case, there are $N$ comparisons if the first 2 values are the 2 largest values.     
    - In a heap, you make comparisons when:
        1. when you move nodes around, you always need to compare with the 2 child nodes for every node you move. In the worst case, you will make $2 * \log N$ comparisons with every pop. 
        2. when you build the heap, you compare every item with at least 2 other items, because a heap is a complete binary tree
        3. This gives us $2N + 2 \log N$ total comparisons
        4. You cannot really optimise more, because of how heaps work

### Min-Heap (Size 2)

- Maintain a Min-Heap with a fixed capacity of 2.
- The "Min" property ensures the smallest of our "top candidates" is always at the root.

- Process:
    - Seed the heap with the first two elements of the list.
    - For every subsequent element, compare it to the heap root.
    - If the element is larger than the root, replace the root and "heapify down" (swap it with the other element if it's smaller).

- Complexity:
    - Initialization: O(1) (Size is constant)
    - Iteration: O(N) to look at every element.
    - Heapify: Since the tree height is log(2), it costs effectively O(1) per update.
    - Overall Time: O(N)

    - Space: O(1) (We only store 2 elements regardless of how large N is).

- Constant Computational Overhead:
    - For every item, you make 1 comparison against the root.
    - If the item is larger, you make 1 additional comparison to re-order the 2-element heap.
    - This results in a range of N to 2N comparisons total.

- Memory Performance: Like the 2-pointer solution, this is extremely cache-friendly if implemented using a tiny array, as the "heap" fits entirely in the CPU's fastest registers.

- Comparison to 2-Pointer Solution
    - The 2-Pointer solution is essentially a "manual" implementation of a size-2 Min-Heap.
    - Why it feels different: The 2-pointer code uses if/else logic to maintain order, whereas the Heap uses the heapify property.
    - The "Winner" Argument: - For K=2, the 2-pointer solution is usually preferred for its simplicity.
    - For K > 5, the Min-Heap is vastly superior because the code remains clean, while the 2-pointer approach would require a massive, error-prone chain of if/else statements.

### BST

- Insert all $N$ elements from the list into a BST
- Once the tree is built, the largest value is the rightmost node 
- The second largest is the node you encounter just before reaching that "max" or the rightmost node of the max's left subtree

- Process:
    - For each element in the list, perform a standard BST insertion.
    - Once all $N$ items are inserted, start at the Root.
    - Case A: If the rightmost node has a left subtree, the second largest is the rightmost node of that left subtree.
    - Case B: If the rightmost node is a leaf, the second largest is its parent.

- Complexity:
    - Building the Tree (Average Case): $O(N \log N)$ if the input is random, as each insertion takes $O(\log N)$
    - Building the Tree (Worst Case): $O(N^2)$ if the input is already sorted. The tree becomes a "skewed" line (essentially a linked list)
    - Finding the Value: $O(h)$ where $h$ is the height of the tree ($O(\log N)$ average, $O(N)$ worst)
    - Overall Time: $O(N \log N)$ average
    - Space: $O(N)$ to store the nodes and pointers

- Constant Computational Overhead:
    - Pointer Overhead: Unlike the array-based heap, a BST requires storing at least two pointers (left and right) per node. This uses significantly more RAM.
    - Comparison Count: Every insertion requires multiple comparisons to find the correct "hole" in the tree. By the time you find the second largest, you've performed far more comparisons than the 2-pointer or Heap methods.
    - No Early Exit: You must wait for the tree to be fully constructed (or at least partially balanced) before you can reliably find the second largest

- Addition: What happens if input is sorted?
    - The tree becomes a line and performance tanks to $O(N^2)$
    - Can use Self-Balancing BST (like an AVL or Red-Black tree) to keep insertion at $O(\log N)$

- Addition: How does your BST handle the list [10, 10, 8]?
    - They need to decide if they store duplicates in the left or right child, or if they keep a "count" at each node. This determines if they find the second unique value or not.

- Addition: What happens when you need the 3rd largest value instead?
    - BST is not efficient past 2nd largest.
    - You need to implement some sort of reverse in-order traversal, AND you need O(N) space to store the tree

### Set

- The Strategy: Deduplicate the list using a Hash Set, then find the max of the remainder.

- Complexity:
    - Time: O(N) average to build the set.
    - Space: O(N) worst-case to store unique values.
- Why use it?
    - It is the most robust way to handle "Unique Second Largest."
    - It simplifies the logic: once duplicates are gone, any selection algorithm works without "if x == max1" edge cases.

- Downsides:
    - High memory overhead compared to 2-pointers.
    - Performance depends on the quality of the Hash Function. If many elements collide, time complexity can degrade to O(N^2) in some languages/implementations.
        - Why? Because what happens is that the 2 elements with the same hash will be stored in a linked list. 
        - In the worst case with $N$ elements colliding, you iterate over your original list with $O(N)$ time, and then for each element $N$, you iterate over the linked list again to check for equality, taking another $O(N)$, leading to $O(N^2)$
